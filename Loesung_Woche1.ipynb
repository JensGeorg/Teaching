```r
test<-sample(1:50,100,replace=T) # Testvektor
```

```r
test
```

```r
# Minimum Skript

mi<-test[1]
for(i in test){
  if(i < mi){
    mi<-i
  }
}
mi
```

```r
# Maximum Skript

ma<-test[1]
for(i in test){
  if(i > ma){
    ma<-i
  }
}
ma
```

```r
#Haeufigkeit Minimum

mi<-test[1]
number<-0
for(i in test){
  if(i == mi){
    number<-number+1
  }
  if(i < mi){
    mi<-i
    number<-1
  }
}
mi 
number
```

```r

#Haeufigkeit Maximum

ma<-test[1]
number<-0
for(i in test){
  if(i == ma){
    number<-number+1
  }
  if(i > ma){
    ma<-i
    number<-1
  }
}
ma 
number
```

```r
# Welche Zahlen kommen wie oft vor

numbers<-c(0)
names(numbers)<-test[1]

for(i in test){
  id<-as.character(i)
  if(is.na(numbers[id])){
    numbers[id]<-1
  } else {
    numbers[id]<-numbers[id]+1
  }
}
numbers
```

```r
# Bubblesort

sorted_vektor<-test

for(i in 1:(length(sorted_vektor)-1)){
  for(j in 1:(length(sorted_vektor)-i)){
    zahl1<-sorted_vektor[j]
    zahl2<-sorted_vektor[j+1]
    if(zahl1>zahl2){
      sorted_vektor[j+1]<-zahl1
      sorted_vektor[j]<-zahl2
    }
  }
}
sorted_vektor
```
